```python
import json
import pandas as pd

# open the JSON and load in the content
with open('nba_player_data.json', 'r',encoding="utf8") as file: # "temp.json" is the name of the json we want to rea din. Change this to the name of your JSON file
    data = json.load(file)

# Check/Get a Look at the JSON structure
print(json.dumps(data, indent=3)) 
```

```python
headers = headers[0].to_list() # Make the headers into a list 
headers
```

```python
# Now we can simply rename by setting the columns equal due to them being in the right order and size 
df.columns = headers
```

and Voila! You have a workable dataframe to use!

# Method 3: [Beautiful Soup](https://www.crummy.com/software/BeautifulSoup/bs4/doc/)

This one is for HTML sites where if you use Inspect Element on the site and look at the HTML and can see the data you want to scrape, it will work there. This method allows you to scrape the whole sites data rather than just the table content, but for the intro to webscrapping element, we will stick with tables to make it a little more simple to wrap heads around since tables are common elements we are used to navigate rather than raw HTML. If you have experiance in raw HTML, you will be able to see what tags you need to search up rather than 'table' like we will here.

We will stick to using the Medical Specialty Wikipedia page to get the table of specialties!

```python
# Install the library. Uncomment to install if needed
#!pip install BeautifulSoup4

# To use it, import the module 
from bs4 import BeautifulSoup 

# We will use pandas as well to do some cleaning!
import pandas as pd
```

Before we really get started we need to "request" the HTML from the URL we want to use. We will use the built in requests library to do so!

```python
import requests

url = "https://en.wikipedia.org/wiki/Medical_specialty" # Put the url of the site you want to parse 

site_html = requests.get(url) ## Actually gets the HTML from the URL
```

Its a simple as that to get the HTML without needit to copy paste things yourself! From there, we can actually do the web scrapping part.

We first need to initialize BeautifulSoup to use the parse we want so it know what to use to look at the html but also the html we want it to look at. "html.parser" is what we will use to parse as mentioned and is how we will call the built in method. If you happen to use another parser, you would replace it with it (ex. "html5lib" or "lxml") and "site_html" is what we will call for the html!

```python
soup = BeautifulSoup(site_html.content, "html.parser")

soup.prettify ## take a look at the html and ensure it looks somewhat right
```

```python
soup.find_all('table')[4] #Change the index to be the table on the page you want. Ours is 4 here!
```

Now that we know what element to use, we can create a variable to hold the table! 

```python
specialty_table = soup.find_all('table')[4]

specialty_table
```

Now we have the tables HTML! You can use pandas read_html to do the rest for us. It does need to be cast into a string object so we will wrap it in a str() to make it a string. 

There are also different methods within BeautifulSoup to also find elements you need and get shown in later sections but pandas will be the go to for now!

```python
specialty_df = pd.read_html(str(specialty_table))[0] # Convert the HTML into a string and set the index after to be 9 so that it pull the table

specialty_df
```

And now we have a table to do something with!

# Method 4: [Selenium](https://www.selenium.dev/documentation/)
#### **WARNING: Make sure you set this up in a virtual environment!**

```python
# Uncomment to install to ensure this doesn't install if you hit run all or something :)
# **WARNING: Make sure you set this up in a virtual environment!**
#!pip install selenium
```

```python
# Gather all the imports! Uncomment the Service if you want to play
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup 
import json
import pandas as pd

# Set up Selenium with Chrome
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode (no UI)
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Create a browser instance using Chrome
driver = webdriver.Chrome()

# Placeholder for our content since we aren't exporting it just yet
soup = "" 

try:
    # Specify the URL the driver will be getting the data from
    driver.get("https://api.dineoncampus.com/v1/location/587124593191a200db4e68af/periods/677fef2ce45d4306446fa1c9?platform=0&date=2025-1-21")
    
    # Set a wait time if things need to load. This is optional but here the website took a little bit of time to call its data so we set it here!
    driver.implicitly_wait(60)
    
    # Scrape the response/HTML
    page_source = driver.page_source
    
    # Use BeautifulSoup to parse the HTML
    soup = BeautifulSoup(page_source, "html.parser")
    
    # Print the HTML to see if things worked well!
    print(soup.prettify())

finally:
    # Close the browser. Good practice to do this!
    driver.quit()
```

It looks like it went well so now we have the HTML with the data but the data is contained in a JSON. We can clean it up to get just the JSON and move from there. In this case, its a little wasier to extract the data since it is inbetween a certain tag (< pre>) making it something we can easily extract! Sometimes you may need to do more intensive cleaning if the tags aren't easy using string methods in different iterations or other processing libraries. 

```python
# Using beautifulSoup, find the "pre" tag (there is only one here) and strip it
# Json loads will ensure it is formatted correctly in later steps!
json_content = json.loads(soup.find("pre").text.strip())

json_content
```

Now we have the data, but as a JSON! We will then use pandas to get the data we need and put it in a dataframe so use however is needed! 

Due to the nesting nature of the arrays, we will need to utilize multiple normalization calls and merge. 

```python
items = []
for category in json_content["menu"]["periods"]["categories"]:
    for item in category["items"]:
        item["type"] = category["name"]  # Add the type/category name
        items.append(item)

#Get the main items
df = pd.json_normalize(items)

#get the nutrient information
nutrients = pd.json_normalize(items, "nutrients", ["id", "name"], record_prefix='nutrient_')

#Get the filters
filters = pd.json_normalize(items, "filters", ["id", "name"], record_prefix="filter_")   
```

```python
nutrients
```

```python
filters
```

Looking at what we have so far, it has each type of allergen as a row for each item. Same thing with nutrition where each row represents one nutritional element of an item. We ideally want it so each item is a row and the columns housing the nutritional values. All it needs is to be reshaped! 

```python
# reshape the nutrients and filters to be more in the format we want to merge with so each item is one row
nutrients_df = nutrients.pivot(index=["id", "name"], columns="nutrient_name", values="nutrient_value").reset_index()

# We will use pivot table here so that way we can mark an allergen as true or false based on if it exists within the list of filters.
filters_df = filters.pivot_table(index=["id", "name"], columns="filter_name", values="filter_type",aggfunc=lambda x: True ).reset_index()

# Fill the nulls for those that didn't list the other existing allergens/labels
filters_df = filters_df.fillna(False)
```

```python
filters_df
```

```python
nutrients_df
```

Now that everything is in order, we can put it all together into a big DataFrame!

```python
# Merge into one nice dataframe!
df = df.merge(nutrients_df, on="id", how="left").merge(filters_df, on="id", how="left")

df
```

One day isn't going to have every menu item possible so you may want to scrape multiple days. If the url you are using has the parameters for how it is searching for the data, you can easily maipulate the URL to work within a look and take and array of values to change the search parameters. 

You could play around with location ID's as well but we will just play with date here as its a common one! We will start by specifying an array for the dates we want to scrape. 

```python
from datetime import datetime, timedelta

# Generate a range of dates
start_date = datetime(2025, 1, 21)  # Start date
end_date = datetime(2025, 1, 25)    # End date
delta = timedelta(days=1) # Number of days to increment by

# Create a list of dates in the correct format for the URL (YEAR-MONTH-DAY)
dates = [(start_date + i * delta).strftime('%Y-%m-%d') for i in range((end_date - start_date).days + 1)]

# Get a look at some of them to ensure they are in the right format. 
dates[1:5] #splicing here incase you want to make the array longer :)
```

We can then wrap what we did previously into a for loop to iterrate through the dates, add them to the URL and do the same old scrape! This time we will also add in the Pre-cleaning to leave us with a final csv!

```python

for date in dates: 
    # Open Chrome Browser
    driver = webdriver.Chrome()
    
    try:
        # Specify the URL the driver will be getting the data from
        driver.get(f"https://api.dineoncampus.com/v1/location/587124593191a200db4e68af/periods/677fef2ce45d4306446fa1c9?platform=0&date={date}")
        print(f"Gathering data for {date}!")
        
        # Set a wait time if things need to load. This is optional but here the website took a little bit of time to call its data so we set it here!
        driver.implicitly_wait(200)
        
        # Scrape the response/HTML
        page_source = driver.page_source
        
        # Use BeautifulSoup to parse the HTML
        soup = BeautifulSoup(page_source, "html.parser")
        
        json_content = json.loads(soup.find("pre").text.strip())

        items = []
        for category in json_content["menu"]["periods"]["categories"]:
            for item in category["items"]:
                item["type"] = category["name"]  # Add the type/category name
                items.append(item)

        #Get the main items
        df = pd.json_normalize(items)

        #get the nutrient information
        nutrients = pd.json_normalize(items, "nutrients", ["id", "name"], record_prefix='nutrient_')

        #Get the filters
        filters = pd.json_normalize(items, "filters", ["id", "name"], record_prefix="filter_")   

        nutrients_df = nutrients.pivot(index=["id", "name"], columns="nutrient_name", values="nutrient_value").reset_index()

        # We will use pivot table here so that way we can mark an allergen as true or false based on if it exists within the list of filters.
        filters_df = filters.pivot_table(index=["id", "name"], columns="filter_name", values="filter_type",aggfunc=lambda x: True ).reset_index()

        # Fill the nulls for those that didn't list the other existing allergens/labels
        filters_df = filters_df.fillna(False)

        df = df.merge(nutrients_df, on="id", how="left").merge(filters_df, on="id", how="left")

        #Export the CSV so that way you can use it and not need to keep scraping!
        df.to_csv(f"{date}.csv",index = False)
        
    finally:
        # Close the browser. Good practice to do this!
        driver.quit()
```
